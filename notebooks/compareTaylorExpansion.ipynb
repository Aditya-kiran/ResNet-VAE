```python
import numpy as np

%pylab inline
```

```python
x = np.linspace(-0.8,0.8,1000)
y1 = np.log(1-x)
y2 = -x -x**2

hold(True)
plot(x,y1,'r',label='exact')
plot(x,y2,'g',label='approx')
axvline(x=0.68)
legend()
```

```python
x = np.linspace(-0.8,0.8,1000)
y1 = np.log(1+x)
y2 = x -x**2

hold(True)
plot(x,y1,'r',label=r'$\log(1+x)$')
plot(x,y2,'g',label=r'$x - x^2$')
axvline(x=-0.68)
legend()
```
