```python
from pandas import Series
from matplotlib import pyplot as plt
series1= Series.from_csv("../code/reg_data.csv", header=0, index_col=[0])
series2= Series.from_csv("../code/run_0.1_40_1_resnet_8_20000_0.01_1_tensorboard-tag-loss.csv", header=0, index_col=[0])
series3= Series.from_csv("../code/run_0.0_40_1_resnet_8_20000_0.05_123_tensorboard-tag-loss.csv", header=0, index_col=[0])
series4= Series.from_csv("../code/run_0.1_40_1_resnet_8_20000_0.05_123_tensorboard-tag-loss.csv", header=0, index_col=[0])
# Tail-rolling average transform
rolling1= series1.rolling(window=10)
rolling2= series2.rolling(window=10)
rolling3= series3.rolling(window=10)
rolling4= series4.rolling(window=10)

rolling1_std = rolling1.std()
rolling2_std = rolling2.std()
rolling3_std = rolling3.std()
rolling4_std = rolling4.std()
# plot original and transformed dataset
# series.plot()
rolling1_std.plot(color='Green')
rolling2_std.plot(color='red')
rolling3_std.plot(color='blue')
rolling4_std.plot(color='purple')
plt.show()
```

```python
df1 = pd.DataFrame({'first_order_no_reg':rolling1_std.values, 'first_order_with_reg':rolling2_std.values, \
                   'second_order_no_reg':rolling3_std.values, 'second_order_with_reg':rolling4_std.values})
df1.to_csv("../code/std.csv")
# df2.to_csv("../code/std.csv")
```
